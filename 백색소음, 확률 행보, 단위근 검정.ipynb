# 여기는 확률 행보

```python
import numpy as np
```

```python
probability = [0.5, 0.5]
start = 0
rand_walks=[start]
```

```python
rand_point = np.random.random(300)
down_probability = rand_point < probability[0]
up_probability = rand_point >= probability[1]
```

```python
for down, up in zip(down_probability, up_probability):
    rand_walks.append(rand_walks[-1] - down+up)
```

```python
plt.plot(rand_walks)
plt.show()
```

# 단위근 검정

```python
import numpy as np
import matplotlib.pyplot as plt
init = np.random.normal(size=1, loc=3)
e = np.random.normal(size=1000, scale= 3)

y = np.zeros(1000)
z = np.zeros(1000)
y[0] = init
z[0] = init

for t in range(1, 1000):
    y[t] = 3 + 1*y[t-1] + e[t]
    z[t] = 3 + 0.1*z[t-1] + e[t]
fig, ax = plt.subplots(1, 2)
ax[0].plot(y)
ax[0].set_title("Non-Stationary")
ax[1].plot(z)
ax[1].set_title("Stationary")
plt.show()
```
